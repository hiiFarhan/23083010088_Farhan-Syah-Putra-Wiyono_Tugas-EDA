```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
print("Kolom numerik yang tersedia:", numeric_columns)

# Gantilah 'hotel_name' dengan nama kolom yang sesuai jika diperlukan, misalnya 'location' atau 'region'
group_column = 'name'  # Asumsi ini adalah kolom yang ingin Anda gunakan untuk pengelompokan

# Memeriksa apakah kolom grup benar ada di dalam dataset
if group_column not in df.columns:
    print(f"Kolom '{group_column}' tidak ditemukan di dataset. Gantilah dengan kolom yang tepat.")
    
# Konfigurasi subplot untuk menampilkan grafik bar
n_cols = 1
n_rows = (len(numeric_columns) + n_cols - 1) // n_cols 

# Simpan data Top 10 dalam dictionary
top_10_lists = {}

# Loop melalui setiap variabel numerik untuk menghitung Top 10
for var in numeric_columns:
    top_10 = df.groupby(group_column)[var].sum().reset_index().sort_values(by=var, ascending=False).head(10)
    top_10_lists[var] = top_10

# Plot bar chart untuk setiap variabel numerik
plt.figure(figsize=(10, 6 * n_rows))

for i, var in enumerate(numeric_columns):
    plt.subplot(n_rows, n_cols, i + 1)
    top_10 = top_10_lists[var]
    plt.bar(top_10[group_column], top_10[var], color='skyblue', edgecolor='black')
    plt.xlabel('Hotel Name')  # Sesuaikan label x-axis jika perlu
    plt.ylabel(f'{var}')
    plt.title(f'Top 10 {var}')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()  
plt.show()
```

```python
variable_pairs = [
    ('interest', 'penalty'),  # Contoh pasangan kolom
    # Tambahkan pasangan variabel lain jika diperlukan
]

# Asumsikan ada kolom seperti 'name' yang mewakili tempat atau nama hotel
group_column = 'name'  # Ganti dengan kolom yang tepat seperti 'hotel_name'

# Cek apakah kolom untuk grouping ada dalam dataset
if group_column not in df.columns:
    print(f"Kolom '{group_column}' tidak ditemukan di dataset. Gantilah dengan kolom yang benar.")

# Menghitung total interest dan penalty untuk setiap lokasi
grouped_df = df.groupby(group_column)[['interest', 'penalty']].sum().reset_index()

# Memilih 10 lokasi teratas berdasarkan jumlah interest
top_locations = grouped_df.nlargest(10, 'interest')[group_column]

# Pengaturan grid plot
n_cols = 1
n_rows = len(variable_pairs)

plt.figure(figsize=(10, 6 * len(variable_pairs)))  

# Loop untuk membuat plot berdasarkan pasangan variabel
for i, (interest_var, penalty_var) in enumerate(variable_pairs):
    plt.subplot(n_rows, n_cols, i + 1)
    
    # Filter data hanya untuk 10 lokasi teratas
    top_data = grouped_df[grouped_df[group_column].isin(top_locations)]
    
    # Plot untuk kolom 'interest' dan 'penalty'
    plt.plot(top_data[group_column], top_data[interest_var], label='Interest', marker='o')
    plt.plot(top_data[group_column], top_data[penalty_var], label='Penalty', marker='s')

    plt.xlabel('Location')  # Sesuaikan label sesuai konteks
    plt.ylabel('Amount')
    plt.title(f'{interest_var.capitalize()} vs {penalty_var.capitalize()} by {group_column.capitalize()}')
    plt.xticks(rotation=45, ha='right')
    plt.legend()

plt.tight_layout()
plt.show()
```
